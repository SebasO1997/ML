## Distribucion de datos e Histogramas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os 
```

```python
datos_prueba = np.arange(1,11,1)
datos_prueba = datos_prueba.tolist()
datos_prueba
```

#### __Histrograma__
Es una grafica que permite ver la frecuencia de los valores dentro de los datos; además, es una representacion aproximada de la distrubucion de los datos 

```python
plt.hist(datos_prueba, color='pink', ec='black') ## ec: color de linea
plt.show()
```

Que pasa si adiciono más numeros a lista dentro del rango ?

```python
datos_prueba.append(1)
```

```python
plt.hist(datos_prueba, color='pink', ec='black') ## ec: color de linea
plt.show()
```

Se pueden juntar valores por cubetas y lograr mostrarlos de 2 en 2 o n en n; agregando bins = 

```python
plt.hist(datos_prueba,bins=5, color='pink', ec='black') ## ec: color de linea
plt.show()
```

#### __Generar arreglos con datos aleatorios__

```python
np.random.rand(10)
```

#### Arreglo con distribucion normal

```python
np.random.normal(loc = 100, scale=10, size=1000) # loc: modifico la media, scale: modifca la desviacion estandar
```

```python
print(np.random.normal(loc = 100, scale=10, size=1000).mean())
print(np.random.normal(loc = 100, scale=10, size=1000).std())
```

#### Tres distribucion importantes 
* Normal
* Uniforme
* Exponencial 

```python
### NORMAL  
## Media en el 0 
datos_d_normal = np.random.normal(size=10000)
plt.hist(datos_d_normal, color='pink', ec='w')
plt.title('Distribucion normal')
plt.show()

## Uniforme
datos_d_uniforme = np.random.uniform(size=10000)
plt.hist(datos_d_uniforme, color='skyblue', ec='w')
plt.title('Distribucion uniforme')
plt.show()

## Exponencial
datos_d_exponencial = np.random.exponential(size=10000)
plt.hist(datos_d_exponencial, color='green', ec='w')
plt.title('Distribucion Exponencial')
plt.show()
```
