## visualizacion con de Ralaciones Estadisticas con Python

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
pinguinos = sns.load_dataset('penguins')
```

```python
pinguinos
```

### Primera grafica 
* Replot: busca relaciones en los datos

#### Contexto de una grafica
* Estilo: ajuste tema principal, sns.set_theme(style = '') style = [white, dark, darkgrid, ticks]
* context: Ajusta el tamaño de los textos: sns.set_theme(style = '', context= '') context=[paper, notebook, talk , poster]   
Para personalizar más la grafica se debe crear una instancia de la misma 



```python
sns.set_theme(style ='dark', context='talk')
relacional = sns.relplot(data=pinguinos, height=8, aspect=1.2, markers=['s', '.', 'P','d']) # height: tamaño de la grafica, aspect: ancho de la grafica 
relacional.set(title='Analisis de pinguinos')
relacional.set_ylabels('Variables')
relacional.set_xlabels('Numero de pinguino')

# plt.grid(False) # ambas librearias trabajan sobre el mismo lienzo
plt.show()
```

##  Diagramas de Dispersion 

```python
## Grafica simple 
sns.relplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm')
```

```python
## Grafica Matiz y tamaño fijo
sns.relplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm', hue='body_mass_g', s=100)
```

```python
sns.relplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm', hue='species', s=100)
```

```python
sns.relplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm', hue='species', row='sex', size='body_mass_g', sizes=(100,500), style='sex')
```

```python
sns.relplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm', hue='species', col='island',row='sex', size='body_mass_g', sizes=(100,500), style='sex')
```

## Graficas de lineas

```python
sns.relplot(data=pinguinos, x='species', y='body_mass_g', kind='line', hue='sex', style='sex')
plt.show()
```

```python
sns.relplot(data=pinguinos, x='bill_depth_mm', y='flipper_length_mm', kind='line', hue='species', style='species')
plt.show()
```

## Combinando graficos 
### Lineas y dispersion

```python
figure, axis =  plt.subplots(1,2, figsize=(15,5))
sns.lineplot(data=pinguinos, x='species', y='body_mass_g', hue='sex', style='sex', ax=axis[0])
sns.scatterplot(data=pinguinos, x='bill_length_mm', y='flipper_length_mm', hue='species', s=100, ax=axis[1])
plt.show()
```
