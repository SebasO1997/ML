## Diagramas de Caja (BOXPLOTS) y Datos anomalos(outliers) con la prueba de __TUKEY__ en python

Cuando se pide 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
edades = np.random.randint(10,90, size=(10))
edades
```

```python
plt.boxplot(edades)
plt.grid(True)
plt.yticks(np.arange(0,110, 5))
plt.show()
```

#### Calculo de los cuartiles que forman el boxplot

```python
(np.min(edades), np.quantile(edades,0.25), np.median(edades), np.quantile(edades,0.75), np.max(edades)) # primer cuartil 
```

### Datos anormales/ atipicos/ ouliers
#### __Pruebas de Tukey__

```python
edades[-1] = 110
edades[-1 ]= 150
edades[-1 ]= 200
```

```python
plt.boxplot(edades)
plt.grid(True)
plt.yticks(np.arange(0,210, 5))
plt.show()
```
