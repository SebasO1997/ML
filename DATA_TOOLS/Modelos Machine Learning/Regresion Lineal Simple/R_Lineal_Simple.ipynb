### Creacion del Modelo
#### Regresion Lineal Simple

```python
from sklearn import linear_model # Modulo de modelos lineales

## Creacion de un creador de modelos de regresion lineal 
regresion  = linear_model.LinearRegression()
```

```python
## Preparacion de los datos 
## este modelo solo tiene una variables objetivo (a predecir) y una sola de entrada

## Ajuste del formato de los datos de entrada
vector_horas = datos_ingresos['horas'].values.reshape((-1,1))
                            ##Accedo. Reajuste de la forma((hasta el ultimo elemento(-1), agrupamiento(1)))
vector_horas
```

```python
## Creacion del modelo
 ## corresponde una linea recta
modelo = regresion.fit(vector_horas, datos_ingresos['ingreso']) ## fit(var Ind, var Dep)
```

```python
print(f' Interseccion (b) {modelo.intercept_}')
print(f' Pendiente (m) {modelo.coef_}')
```

La pediente indica que por cada hora trabajado el salario aumenta aproximadamente 3k pesos

### Predicciones 

```python
entrada = [[39],[40],[43],[44]]

modelo.predict(entrada)
```

regresa un array con el potencial salario de la personas en funcion de sus horas trabajadas 

```python
plt.scatter(entrada, modelo.predict(entrada), color= 'red')
plt.plot(entrada, modelo.predict(entrada), color = 'black')
plt.scatter(datos_ingresos['horas'], datos_ingresos['ingreso'])

plt.ylabel('Entrada')
plt.xlabel('Horas semanales trabajadas')
plt.show()
```
